```python
# 1 - Tentar vincular dados do IBGE com os dados de imóveis
errado = dados_ibge.columns[1]
dados_ibge.columns[:21]
```

```python
dados_ibge = dados_ibge.iloc[:,:21]
dados_ibge.rename(columns={errado: 'Cod_Grandes_Regioes'}, inplace=True)
print(dados_ibge.shape)
dados_ibge.head()
```

Vamos ver qual coluna pode ser eliminada:

```python
dados_ibge['Cod_Grandes_Regioes'].value_counts()  # todos são no sudeste, as duas colunas podem ser retiradas
```

```python
dados_ibge['Cod_UF'].value_counts()  # todos da mesma UF (pode tirar Cod_UF e Nome_da_UF)
```

```python
dados_ibge['Cod_meso'].value_counts()  # todos do mesmo MESO
```

```python
dados_ibge['Cod_micro'].value_counts()  # todos do mesmo micro
```

```python
dados_ibge['Cod_RM'].value_counts()  # todos do mesmo MR
```

```python
dados_ibge['Cod_municipio'].value_counts()  # todos do mesmo municipio
```

```python
dados_ibge['Cod_distrito'].value_counts()  # varios distritos
```

```python
dados_ibge['Cod_bairro'].value_counts()  # apenas um bairro
```

Podemos tirar as colunas de: GRANDES REGIOES,UF, MESO, MICRO, RM, MUNICIPIO e bairro

```python
dropar_colunas = ['Cod_Grandes_Regioes', 'Nome_Grande_Regiao', 'Cod_UF',
       'Nome_da_UF ', 'Cod_meso', 'Nome_da_meso', 'Cod_micro', 'Nome_da_micro',
       'Cod_RM', 'Nome_da_RM', 'Cod_municipio', 'Nome_do_municipio','Cod_bairro', 'Nome_do_bairro']
dados_ibge.drop(dropar_colunas, axis=1, inplace=True)
dados_ibge.head()
```

Além disso, vamos retirar os nomes, guardar numa variável e usar apenas os números

```python
distritos = dados_ibge.groupby('Nome_do_distrito')['Cod_distrito'].unique()
distritos
```

```python
subdistritos = dados_ibge.groupby('Nome_do_subdistrito')['Cod_subdistrito'].unique()
subdistritos
```

```python
dados_ibge.drop(['Nome_do_distrito', 'Nome_do_subdistrito'], axis=1, inplace=True)
```
